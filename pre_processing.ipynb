```python
import pandas as pd

# Load the provided CSV file
file_path = 'data/shopping_trends.csv'
shopping_data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure and contents
shopping_data.head()
```

```python
# Selecting the specified columns
filtered_df = shopping_data[['Age', 'Gender', 'Item Purchased', 'Category', 'Season', 'Payment Method', 'Shipping Type']]

# Renaming the columns
filtered_df.rename(columns={'Item Purchased': 'Item_Purchased', 
                        'Payment Method': 'Payment_Method',
                        'Shipping Type': 'Shipping_Type'}, inplace=True)
```

```python
# 檢查缺失值
missing_values = filtered_df.isnull().sum()
print(missing_values)

# 考慮數值型欄位進行異常值檢查，例如 'Age'
numeric_columns = ['Age']
outliers = filtered_df[numeric_columns].describe()
print(outliers)
```

```python
# Path for the new CSV file
new_csv_path = 'data/preprocessed_shopping_trends.csv'

# Saving the filtered dataframe to a new CSV file
try:
    filtered_df.to_csv(new_csv_path, index=False)
    success = True
except Exception as e:
    success = False
    error_message = str(e)

new_csv_path if success else error_message
```
